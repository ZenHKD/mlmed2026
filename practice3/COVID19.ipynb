```python
import pandas as pd
import matplotlib.pyplot as plt 

log1 = pd.read_csv("phase1_training_log.csv")

epoch1            = log1['epoch'].tolist()

train_loss1       = log1['train_loss'].tolist()
train_seg_loss1   = log1['train_seg_loss'].tolist()
train_cls_loss1   = log1['train_cls_loss'].tolist()
train_dice1       = log1['train_dice'].tolist()
train_iou1        = log1['train_iou'].tolist()
train_acc1        = log1['train_acc'].tolist()

val_loss1         = log1['val_loss'].tolist()
val_seg_loss1     = log1['val_seg_loss'].tolist()
val_cls_loss1     = log1['val_cls_loss'].tolist()
val_dice1         = log1['val_dice'].tolist()
val_iou1          = log1['val_iou'].tolist()
val_acc1          = log1['val_acc'].tolist()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 18), dpi=200, sharex=True)

# 1. Total Loss
axes[0].plot(epoch1, train_loss1, label='Train Loss', color='blue')
axes[0].plot(epoch1, val_loss1, label='Validation Loss', color='orange')
axes[0].set_ylabel('Total Loss')
axes[0].set_title('Total Loss')
axes[0].grid(True)
axes[0].legend()

# 2. Segment Loss
axes[1].plot(epoch1, train_seg_loss1, label='Train Segment Loss', color='green')
axes[1].plot(epoch1, val_seg_loss1, label='Validation Segment Loss', color='red')
axes[1].set_ylabel('Segment Loss')
axes[1].set_title('Segmentation Loss')
axes[1].grid(True)
axes[1].legend()

# 3. Classify Loss
axes[2].plot(epoch1, train_cls_loss1, label='Train Classify Loss', color='purple')
axes[2].plot(epoch1, val_cls_loss1, label='Validation Classify Loss', color='brown')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Classify Loss')
axes[2].set_title('Classification Loss')
axes[2].grid(True)
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 18), dpi=200, sharex=True)

# 1. Dice
axes[0].plot(epoch1, train_dice1, label='Train Dice', color='blue')
axes[0].plot(epoch1, val_dice1, label='Validation Dice', color='orange')
axes[0].set_ylabel('Dice')
axes[0].set_title('Dice')
axes[0].grid(True)
axes[0].legend()

# 2. IoU
axes[1].plot(epoch1, train_iou1, label='Train Segment IoU', color='green')
axes[1].plot(epoch1, val_iou1, label='Validation Segment IoU', color='red')
axes[1].set_ylabel('IoU')
axes[1].set_title('IoU')
axes[1].grid(True)
axes[1].legend()

# 3. Classification Accuracy
axes[2].plot(epoch1, train_acc1, label='Train Classify Accuracy', color='purple')
axes[2].plot(epoch1, val_acc1, label='Validation Classify Accuracy', color='brown')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Accuracy')
axes[2].set_title('Classification Accuracy')
axes[2].grid(True)
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
log2 = pd.read_csv("phase2_training_log.csv")

epoch2              = log2['epoch'].tolist()

train_loss2         = log2['train_loss'].tolist()
train_lung_loss2    = log2['train_lung_loss'].tolist()
train_inf_loss2     = log2['train_inf_loss'].tolist()
train_cls_loss2     = log2['train_cls_loss'].tolist()
train_lung_dice2    = log2['train_lung_dice'].tolist()
train_inf_dice2     = log2['train_inf_dice'].tolist()
train_acc2          = log2['train_acc'].tolist()

val_loss2           = log2['val_loss'].tolist()
val_lung_loss2      = log2['val_lung_loss'].tolist()
val_inf_loss2       = log2['val_inf_loss'].tolist()
val_cls_loss2       = log2['val_cls_loss'].tolist()
val_lung_dice2      = log2['val_lung_dice'].tolist()
val_inf_dice2       = log2['val_inf_dice'].tolist()
val_acc2            = log2['val_acc'].tolist()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 12), dpi=200, sharex=True)

# 1. Total Loss
axes[0][0].plot(epoch2, train_loss2, label='Train Loss', color='blue')
axes[0][0].plot(epoch2, val_loss2, label='Validation Loss', color='orange')
axes[0][0].set_ylabel('Total Loss')
axes[0][0].set_title('Total Loss')
axes[0][0].grid(True)
axes[0][0].legend()

# 2. Segment Lung Loss
axes[0][1].plot(epoch2, train_lung_loss2, label='Train Lung Loss', color='green')
axes[0][1].plot(epoch2, val_lung_loss2, label='Validation Lung Loss', color='red')
axes[0][1].set_ylabel('Segment Lung Loss')
axes[0][1].set_title('Segmentation Lung Loss')
axes[0][1].grid(True)
axes[0][1].legend()

# 3. Segment Infection Loss
axes[1][0].plot(epoch2, train_inf_loss2, label='Train Infection Loss', color='purple')
axes[1][0].plot(epoch2, val_inf_loss2, label='Validation Infection Loss', color='brown')
axes[1][0].set_xlabel('Epoch')
axes[1][0].set_ylabel('Segment Infection Loss')
axes[1][0].set_title('Segmentation Infection Loss')
axes[1][0].grid(True)
axes[1][0].legend()

# 4. Classification Loss
axes[1][1].plot(epoch2, train_cls_loss2, label='Train Classify Loss', color='yellow')
axes[1][1].plot(epoch2, val_cls_loss2, label='Validation Classify Loss', color='gray')
axes[1][1].set_xlabel('Epoch')
axes[1][1].set_ylabel('Classify Loss')
axes[1][1].set_title('Classification Loss')
axes[1][1].grid(True)
axes[1][1].legend()

plt.tight_layout()
plt.show()
```
